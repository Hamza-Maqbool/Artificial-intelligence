```python
class AdjNode:
    def __init__(self, value):
        self.vertex = value
        self.next = None


class Graph:
    def __init__(self, num):
        self.V = num
        self.graph = [None] * self.V

    # Add edges
    def add_edge(self, s, d):
        node = AdjNode(d)
        node.next = self.graph[s]
        self.graph[s] = node

        node = AdjNode(s)
        node.next = self.graph[d]
        self.graph[d] = node

    # Print the graph
    def print_agraph(self):
        for i in range(self.V):
            print("Vertex " + str(i) + ":", end="")
            temp = self.graph[i]
            while temp:
                print(" -> {}".format(temp.vertex), end="")
                temp = temp.next
            print(" \n")
    
    # Delete Edge
    def _delete_edge_from_node(self, source, destination):
        prev_node = None
        current_node = self.graph[source]
        while current_node:
            if current_node.vertex == destination:
                if prev_node:
                    prev_node.next = current_node.next
                else:
                    # If it's the first node in the list
                    self.graph[source] = current_node.next
                return
            prev_node = current_node
            current_node = current_node.next

    def get_connected_nodes(self, node):
        if node not in self.graph:
            return []  # Node not found in the graph

        connected_nodes = []
        current_node = self.graph[node]
        while current_node:
            connected_nodes.append(current_node.value)
            current_node = current_node.next

        return connected_nodes

    def are_connected(self, node1, node2):
        if node1 not in self.graph or node2 not in self.graph:
            return False

        visited = set()
        queue = [node1]

        while queue:
            current_node = queue.pop(0)

            # Mark the current node as visited
            visited.add(current_node)

            # Check if the current node is the target node
            if current_node == node2:
                return True

            # Add unvisited neighboring nodes to the queue
            neighbor_node = self.graph[current_node]
            while neighbor_node:
                if neighbor_node.value not in visited:
                    queue.append(neighbor_node.value)
                neighbor_node = neighbor_node.next

        return False

if __name__ == "__main__":
    V = 5

    # Create graph and edges
    graph = Graph(V)
    graph.add_edge(0, 1)
    graph.add_edge(0, 2)
    graph.add_edge(0, 3)
    graph.add_edge(1, 2)

    graph.print_agraph()
    graph._delete_edge_from_node(0, 2)
    graph.print_agraph()
    print("Connected nodes to node 2:", graph.get_connected_nodes(2))
    node1 = 3
    node2 = 2

    if graph.are_connected(node1, node2):
        print(f"Node {node1} and Node {node2} are connected.")
    else:
        print(f"Node {node1} and Node {node2} are not connected.")
```
